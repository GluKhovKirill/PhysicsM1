```python
# CONFIG
PLOTNAME = "Experiment#"
DIR = 'F:\Data\Kurs_1\sem2\physicsLABS\M1\plots' # Путь к директории с графиками
JSON_PATH = 'F:\\Data\\Kurs_1\\sem2\\physicsLABS\M1\\data.json'  # Путь к файлу, выданному 1й программой
DPI = 1200
# --------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
from decimal import Decimal
import decimal
import json
import pandas as pd
import numpy as np


decimal.getcontext().prec=3 # знака после запятой
PI = Decimal('3.142')
G=Decimal('9.81')

get_period = lambda length: 2*PI*( (Decimal(length)/G).sqrt() )

with open(JSON_PATH) as file:
    data = json.loads(file.read())
periods = np.array(data['periods'])

df = pd.DataFrame(periods.transpose(), columns=[f"{PLOTNAME}{str(i)}" for i in range(1,5)])
df = df[(df != -8)]
```

```python
df
```

```python
df.mean() # Среднее значение
```

```python
FACT = [float(get_period(i)) for i in data['lengths']]
abs_error = FACT - df
abs_error # абсолютная ошибка
```

```python
(abs_error / df).mean()*100 # относительная ошибка в процентах
```

```python
def draw_and_save_hist(i='all', draw_all=False):
    if draw_all:
        fig = sns.displot(df, kde=True, color=sns.color_palette("tab10")[:len(df)])
    else:
        fig = sns.displot(df, x=f'{PLOTNAME}{i}', kde=True, color=sns.color_palette("tab10")[i-1])
    plt.grid()
    fig.savefig(f'{DIR}/{PLOTNAME}{i}.svg', format='svg', dpi=DPI)
    fig.savefig(f'{DIR}/{PLOTNAME}{i}.png')
```
